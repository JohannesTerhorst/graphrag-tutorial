```python
import time
import subprocess
from langchain_community.chat_models import ChatOllama
from langchain_core.messages import HumanMessage

# 1. Wir konfigurieren das Modell mit strikten Parametern
windows_ip = subprocess.check_output("ip route list default | awk '{print $3}'", shell=True).decode('utf-8').strip()

chat_model = ChatOllama(
    model="qwen2.5:3b",
    base_url=f"http://{windows_ip}:11434",
    temperature=0.0,    # Keine Kreativität, pure Logik
    num_predict=200     # Maximal 200 Tokens als Antwort
)

# 2. Die Stoppuhr starten
start_time = time.time()

# 3. Den Prompt absenden
response = chat_model.invoke([
    HumanMessage(content="Fasse das Grundgerüst eines from Scratch GraphRAG in einem Satz zusammen. Dockernized GraphRAGs sind nicht erlaubt.")
])

# Stoppuhr anhalten
end_time = time.time()
total_time = end_time - start_time

print(f"Antwort: {response.content}\n")
print("--- EXPERIMENT METRIKEN ---")

# 4. Die Metadaten auslesen (Hier verstecken sich deine Messwerte!)
meta = response.response_metadata

input_tokens = meta.get('prompt_eval_count', 0)
output_tokens = meta.get('eval_count', 0)

print(f"Zeit benötigt:    {total_time:.2f} Sekunden")
print(f"Input-Tokens:     {input_tokens}")
print(f"Output-Tokens:    {output_tokens}")

if total_time > 0:
    print(f"Geschwindigkeit:  {(output_tokens / total_time):.2f} Tokens/Sekunde")
```
