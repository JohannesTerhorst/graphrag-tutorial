```python
import subprocess
from langchain_community.llms import Ollama

# 1. Wir holen uns dynamisch die IP-Adresse von deinem Windows-System aus Sicht von WSL
windows_ip = subprocess.check_output("ip route list default | awk '{print $3}'", shell=True).decode('utf-8').strip()
ollama_url = f"http://{windows_ip}:11434"

print(f"Log: Baue Brücke zu Windows unter {ollama_url} ...\n")

# 2. Wir verbinden das Modell mit der exakten IP
llm = Ollama(model="tinyllama", base_url=ollama_url)

# 3. Der Test
print("Frage ans Modell gesendet... warte auf Antwort...\n")
response = llm.invoke("Hallo! Bitte antworte in einem kurzen Satz: Funktionierst du gerade komplett lokal auf meinem Rechner?")

print(response)
```
